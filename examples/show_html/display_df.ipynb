The DataFrame to be displayed by jupyter notebooks can be adjusted using the ``display_df`` function:

```python
import aaanalysis as aa
df_cat = aa.load_scales(name="scales_cat")
aa.display_df(df_cat, show_shape=True)
```

You can show the shape of the complete DataFrame by setting ``show_shape=True`` and limit visualization by showing only the first ``n_rows`` and ``n_cols``:

```python
aa.display_df(df_cat, show_shape=True, n_cols=2, n_rows=2)
```

Select specific rows and columns using the ``show_only_col`` and ``show_only_row`` parameters:

```python
# Numbering starts with 0
aa.display_df(df_cat, col_to_show="category", row_to_show=1)
```

Limit the number of characters shown in each cell using ``char_limit`` (default=30) 

```python
aa.display_df(df_cat, n_rows=2, char_limit=10)
```

Further style options are ``fontsize`` (default=12), ``max_width_pct`` (default=100), ``max_height`` (default=300): 

```python
aa.display_df(df_cat, n_rows=2, max_width_pct=50, max_height=100)
```
